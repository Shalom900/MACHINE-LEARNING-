```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Generating synthetic data
np.random.seed(0)
n_samples = 100

X = np.random.rand(n_samples, 1) * 10  # Independent variable
y = 2.5 * X.squeeze() + np.random.randn(n_samples) * 2 + 5  # Dependent variable
```

```python
# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Training the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Making predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# Evaluating the model
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

rmse_train = np.sqrt(mse_train)
rmse_test = np.sqrt(mse_test)

mae_train = mean_absolute_error(y_train, y_train_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)

# Printing the evaluation metrics
print(f"Training R-squared: {r2_train:.3f}")
print(f"Test R-squared: {r2_test:.3f}")

print(f"Training Mean Squared Error: {mse_train:.3f}")
print(f"Test Mean Squared Error: {mse_test:.3f}")

print(f"Training Root Mean Squared Error: {rmse_train:.3f}")
print(f"Test Root Mean Squared Error: {rmse_test:.3f}")

print(f"Training Mean Absolute Error: {mae_train:.3f}")
print(f"Test Mean Absolute Error: {mae_test:.3f}")
```

```python
# Getting the coefficients
coefficients = model.coef_
intercept = model.intercept_

# Printing the coefficients
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

```python
# Visualizing the model with training data
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', edgecolor='w', label='Training data')
plt.plot(X_train, y_train_pred, color='red', label='Linear model')
plt.title('Linear Regression Model (Training data)')
plt.xlabel('Predictor')
plt.ylabel('Response')
plt.legend()
plt.show()
```

```python
# Plotting Predictions vs Actual Values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_test_pred, color='green', edgecolor='w', alpha=0.7)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.title('Predicted vs Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()

```
